```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
data = pd.read_csv('Galton.txt',sep='\t')
```

```python
data.columns
```

```python
data.drop(['Family','Gender','Mother','Kids'],axis=1,inplace=True)
```

```python
data.columns
```

```python
X = data['Father']
y = data['Height']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
b1 = ( sum((X_train - np.mean(X_train)) * (y_train - np.mean(y_train))) ) / (sum((X_train-np.mean(X_train))**2))
```

```python
b1
```

```python
b0 = np.mean(y_train ) - b1*np.mean(X_train)
```

```python
D =  np.array(b1 * X_train + b0)
X_train = np.array(X_train)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X_test,y_test)
plt.xlabel('Taille du père (pouce)')
plt.ylabel('Taille du fils (pouce)')
plt.title('Résultat d\'application de la simple régression')
plt.plot(X_train, b1 * X_train + b0,c='green' )
plt.scatter(70,70*b1 + b0, c='b')
plt.scatter(X_train[2],D[2])
```
