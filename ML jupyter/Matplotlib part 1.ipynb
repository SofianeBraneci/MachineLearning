```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x)
```

# Matplotlib part 2 

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)

axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
fig,ax = plt.subplots(nrows= 2, ncols=1 ,figsize=(8,2))
ax[0].plot(x,y)
ax[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')
ax.legend(loc=(0.1,0.1))
```
